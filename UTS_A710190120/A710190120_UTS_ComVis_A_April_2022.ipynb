#### Score: 10, CPMK: 3

Pada gambar `tomat_jeruk.jpg`, buatlah sebuah filter untuk menangkap tomat-nya saja. Tentunya akan lebih mudah kalau kamu membuka dengan colorspace HSV.

```python
# Kode
tomjer = cv2.imread("tomat_jeruk.jpg")
hsv_tomjer = cv2.cvtColor(tomjer, cv2.COLOR_BGR2HSV)

merah_bawah = np.array([0,25,0])
merah_atas  = np.array([20,255,255])

mask1 = cv2.inRange(hsv_tomjer, merah_bawah, merah_atas)

after = cv2.bitwise_and(tomjer, tomjer, mask=mask1)

plt.figure(figsize=[30,10])
plt.subplot(141);plt.imshow(mask1, cmap="gray");plt.title("Hasil mask");
plt.subplot(142);plt.imshow(after);plt.title("hasil extract");
plt.subplot(143);plt.imshow(tomjer[:, :, ::-1]);plt.title("Gambar asli");
```

#### Pertanyaan:
Mengapa untuk tugas semacam ini, colorspace HSV lebih cocok daripada colorspace RGB (atau BGR)?
#### Jawaban:
untuk konversi bgr menjadi rgb dan mempermudah dalam melakukan masking
