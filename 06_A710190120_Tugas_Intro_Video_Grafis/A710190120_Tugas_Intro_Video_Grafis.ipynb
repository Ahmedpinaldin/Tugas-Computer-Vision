```python
## Taruh kodemu di bawah ini. Beri komentar selengkap lengkapnya
gedung_edu = cv2.imread("gedung_edu.jpg", cv2.IMREAD_COLOR)
plt.imshow(gedung_edu[:,:,::-1])
print(gedung_edu.shape)
```

```python
layer_pti_blank = np.zeros(gedung_edu.shape, dtype=np.uint8)
```

```python
#resize gambar
pti = cv2.imread("edu.png", cv2.IMREAD_COLOR)
pti_resize = cv2.resize(pti ,None,fx=0.5, fy=0.5)
print(pti_resize.shape)
```

```python
# meletakan tulisan di bagian pojok kiri atas
top_left_tulisan = np.array([200, 20], dtype=np.uint16)


# meletakan tulisan pada layer
layer_pti = copy.deepcopy(layer_pti_blank)
layer_pti[top_left_tulisan[0]:top_left_tulisan[0]+pti_resize.shape[0], top_left_tulisan[1]:top_left_tulisan[1]+pti_resize.shape[1]  ] = pti_resize
mix = cv2.addWeighted(gedung_edu,1, layer_pti, 0.6 , 0)

plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(layer_pti[:,:,-1], cmap='gray');
plt.subplot(122);plt.imshow(mix[:,:,::-1]);
```

```python
##  Membuat video
fourcc = cv2.VideoWriter_fourcc('m', 'p', '4', 'v')
fps = 20
lebar = gedung_edu.shape[1]
tinggi = gedung_edu.shape[0]

video = cv2.VideoWriter('video_pertama.mp4', fourcc, fps, (lebar,tinggi))

for i in range(144):
    top_left_tulisan = np.array([200 , 10+i], dtype=np.uint16) 
    
    layer_pti = copy.deepcopy(layer_pti_blank)
    layer_pti[top_left_tulisan[0]:top_left_tulisan[0]+pti_resize.shape[0], top_left_tulisan[1]:top_left_tulisan[1]+pti_resize.shape[1]  ] = pti_resize
    mix = cv2.addWeighted(gedung_edu, 1, layer_pti, 0.6 , 0)
    
    text = "Frame : " + str(i)
    cv2.putText(mix, text, (20, 30), cv2.FONT_ITALIC, 1.0, (0, 0, 255), 2, cv2.LINE_AA);
    
    text = "Nama : Ahmed Fisqiya Pinaldin" 
    cv2.putText(mix, text, (20, 60), cv2.FONT_ITALIC, 1.0, (0, 0, 255), 2, cv2.LINE_AA);
    
    text = "NIM : A710190120" 
    cv2.putText(mix, text, (20, 90), cv2.FONT_ITALIC, 1.0, (0, 0, 255), 2, cv2.LINE_AA);
    
    video.write(mix)

video.release()
```

**Pertanyaan:** Apa maksud `fourcc` dalam kodemu? Silakan lihat dokumentasinya di internet, lalu jelaskan kembali dalam bahasa Indonesia.

**Jawaban:** `fourcc` merupakan singkatan dari four Character code, merupakan metode untuk menandai video codec, compression format, serta format warna dan pixel pada video file.

**Pertanyaan:** Apa artinya FPS? Mengapa orang kadang membuat video lebih dari 24 fps?

**Jawaban:** Frame Per Second adalah ukuran kualitas gerakan gambar yang ditampilkan secara berurutan.karena 24 fps merupakan ukuran standard dalam video.
